13. Create a visualise diplay of left footed players with height_cm using histogram

```python
import matplotlib.pyplot as plt
left_footed_players = df.loc[df['preferred_foot'] == 'Left', :]

plt.hist(left_footed_players['height_cm'], bins=10, color='blue', edgecolor='black')
plt.title('Height Distribution of Left-Footed Players')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()
```
